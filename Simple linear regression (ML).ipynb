```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline

df1 = pd.read_csv('Auto.csv')
```

```python
X = df1[['horsepower']]
Y = df1['price']
```

```python
lm = LinearRegression()
```

```python
lm.fit(X,Y)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
Yhat =lm.predict(X)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(X,Y,color = 'r',marker = 'x',s = 75)
plt.title('Horsepower vs Price')
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.grid()

```

```python
#visualization evaluation on the model
import seaborn as sns
%matplotlib inline

width = 8.5
height = 10
plt.figure(figsize=(width,height))
sns.regplot(x = 'horsepower',y = 'price', data =df1,color='r',marker='o')
plt.ylim(0,)
plt.title('Horsepower vs price')
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.show()
```

```python
# residual plot
# visualization evaluation of the model

width = 8.5
height = 8

plt.figure(figsize=(width,height))
sns.residplot(df1['horsepower'],df1['price'])
plt.title('horsepower v price')
plt.show()

```

```python
# calculating the R squared 
lm.score(X,Y)
```

```python
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(Y,Yhat)
print(MSE)
```
