```python
df = pd.read_csv('E:\Datasets\CarPrice_Assignment.csv')
```

```python
df.corr()
```

```python
df1 = pd.DataFrame()

df1 = df[['horsepower','enginesize','curbweight','carwidth','price']]
df1.head()
```

```python
#variables defined

Z = df1[['horsepower','enginesize','curbweight','carwidth']]
y = df1['price']
```

```python
#train and test the model

from sklearn.model_selection import train_test_split

Z_train,Z_test,y_train,y_test = train_test_split(Z,y, test_size = 0.3, random_state = 4)

print(Z_train.shape)
print(Z_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

Lr = LinearRegression()

model = Lr.fit(Z_train,y_train)
```

```python
cofficient = model.coef_
cofficient
```

```python
intercept = model.intercept_
intercept
```

```python
y_predict = model.predict(Z_test)
y_predict[0:5]
```

```python
# plotting distribution for multiple linear regression

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

plt.figure(figsize=(7,7))

ax1 = sns.distplot(y_test,hist = False,color = 'r',label = "Actual values")
sns.distplot(y_predict,hist = False,color ='b',label = 'Predicted values', ax = ax1)

plt.title('Actual values vs Predicted values')
plt.xlabel('Price')
plt.ylabel('Proportion of cars')
plt.legend()
plt.show()
plt.close()
```

```python
Lr.score(Z_train,y_train)
```
