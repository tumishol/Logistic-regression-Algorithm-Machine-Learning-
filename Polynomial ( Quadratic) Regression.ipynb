```python
import pandas as pd
```

```python
df = pd.read_csv('E:\Datasets\CarPrice_Assignment.csv')
```

```python
df.head()
```

```python
df.corr()
```

```python
X = df['curbweight'].values.reshape(-1,1)
y = df['price']
```

```python
X.shape,y.shape
```

```python
df['curbweight'].isnull().sum(),df['price'].isnull().sum()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree = 2)

X_poly = poly.fit_transform(X)
X_poly.shape
```

```python
from sklearn.linear_model import LinearRegression

Lr = LinearRegression()

model = Lr.fit(X_poly,y)
```

```python
coefficient = model.coef_
coefficient
```

```python
intercept = model.intercept_
intercept
```

```python
y_predict = model.predict(X_poly)
y_predict[0:5]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

plt.subplots(figsize=(6.5,6.5))
x_axis = np.arange(1499,4500,0.1)
y_hat = intercept + coefficient[1]*x_axis + coefficient[2]*x_axis**2

plt.scatter(X,y,color = 'b')
plt.plot(x_axis,y_hat,color = 'r',linewidth = 1.75)
plt.title('Curbweight v Price')
plt.xlabel('curbweight')
plt.ylabel('price')
plt.grid()
plt.show()
```

```python
Lr.score(X_poly,y)
```

```python
from sklearn.metrics import mean_squared_error

MSE = mean_squared_error(y,y_predict)
MSE
```
