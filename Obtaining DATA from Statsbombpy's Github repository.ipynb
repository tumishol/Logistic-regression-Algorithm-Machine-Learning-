```python
import requests
import pandas as pd
from pandas import json_normalize
```

```python
competition_id = 43
season_id=3
```

```python
matches = requests.get('https://raw.githubusercontent.com/statsbomb/open-data/master/data/matches/'+str(competition_id)+'/'+str(season_id)+'.json')
matches = matches.json()
```

```python
matches
```

```python
# print all the matches under the competition 2018 FIFA World cup

for match in matches:
    print(match['home_team']['home_team_name'], match['away_team']['away_team_name']+ " " +str(match['home_score'])+ "-" +str(match['away_score'])+ "(match_id is" " "+str(match['match_id'])+")")
```

```python
#print all the france matches in the 2018 FIFA world cup
teamname='France'

for match in matches:
    if match['home_team']['home_team_name']==teamname or match['away_team']['away_team_name']==teamname:
        print(match['home_team']['home_team_name'],match['away_team']['away_team_name']+ " " +str(match['home_score'])+ "-" +str(match['away_score'])+ " " "(match_id is "+str(match['match_id']) +")")
```
