```python
import pandas as pd


df0 = pd.read_csv('E:\\healthcare-dataset-stroke-data.csv') 
df0.head()
```

```python
# converts categorical data into numerical data
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
df0['smoking_status_code'],df0['ever_married_code'],df0['work_type_code'],df0['Residence_type_code'],df0['Gender_code'] = le.fit_transform(df0.smoking_status),le.fit_transform(df0.ever_married),le.fit_transform(df0.work_type),le.fit_transform(df0.Residence_type),le.fit_transform(df0.gender)
df0
```

```python
df1 = pd.DataFrame()

df1['hypertension'],df1['heart_disease'],df1['avg_glucose_level'],df1['bmi'],df1['smoking_status_code'],df1['ever_married_code'],df1['work_type_code'],df1['Residence_type_code'],df1['Gender_code'],df1['stroke'] = df0['hypertension'],df0['heart_disease'],df0['avg_glucose_level'],df0['bmi'],df0['smoking_status_code'],df0['ever_married_code'],df0['work_type_code'],df0['Residence_type_code'],df0['Gender_code'],df0['stroke'] 
```

```python
import numpy as np

mean = df1['bmi'].mean()
df1['bmi'].replace(np.NaN,mean, inplace = True)
df1
```

```python
x = df1[['hypertension','heart_disease','avg_glucose_level','bmi','smoking_status_code','ever_married_code','work_type_code','Residence_type_code','Gender_code']].values
x[0:5]
```

```python
y = df1['stroke'].values
y[0:5]
```

```python
#Normalize the data, this gives zero to mean data and unit variance

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scale = scaler.fit(x).transform(x)
x[0:5]
```

```python
#split data in train and test datasets
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state=4)
x_train.shape
```

```python
#perform the classification using KNN
#Import the KNN library

from sklearn.neighbors import KNeighborsClassifier

#Train the data on the KNN algorithm

k = 5

neigh = KNeighborsClassifier(n_neighbors = k).fit(x_train,y_train)

#Predicting with KNN

yhat = neigh.predict(x_test)
yhat

#accuracy evaluation

from sklearn import metrics

print('accuracy_score for train set:',metrics.accuracy_score(y_train, neigh.predict(x_train)))
print('accuracy_score for test set:', metrics.accuracy_score(y_test,yhat))

```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,yhat)
```

```python
import numpy as np

y_predict_thr3 = np.where(neigh.predict_proba(x_test)[:,1]>0.3,1,0) # threshold is 0.3
y_predict_thr4 = np.where(neigh.predict_proba(x_test)[:,1]>0.4,1,0) # threshold is 0.4
```

```python
def predict_threshold (neigh,x_test,thresholds):
    return np.where(neigh.predict_proba(x_test)[:,1]>thresholds,1,0)
```

```python
for thr in np.arange(0,1.0,0.1):
    y_predict = predict_threshold(neigh,x_test,thr)
    print('Thresholds:',thr)
    print(confusion_matrix(y_test,y_predict))
```
